```python
# Filename: airbnb_app.py
import tkinter as tk
import pandas as pd
import numpy as np

root = tk.Tk()
root.title("Airbnb Data Entry")

# Define fields
fields = ["Name", "Price", "Neighbourhood", "Number of Reviews", "Minimum Nights", "Last Review Date (YYYY-MM-DD)"]
entries = {}
data = []

# Create input fields dynamically
for i, field in enumerate(fields):
    tk.Label(root, text=field).grid(row=i, column=0)
    entry = tk.Entry(root)
    entry.grid(row=i, column=1)
    entries[field] = entry

def add_listing():
    listing = [entries[field].get() for field in fields]
    data.append(listing)
    status_label.config(text="Listing added!")

def show_results():
    # Convert to DataFrame
    df = pd.DataFrame(data, columns=fields)

    # Clean numeric columns
    df["Price"] = df["Price"].replace('[\$,]', '', regex=True).astype(float)
    df["Number of Reviews"] = df["Number of Reviews"].astype(int)
    df["Minimum Nights"] = df["Minimum Nights"].astype(int)

    # Use NumPy for calculations
    avg_price = df.groupby("Neighbourhood")["Price"].mean()
    avg_stay = np.mean(df["Minimum Nights"].values)
    top_reviews = df.iloc[np.argsort(-df["Number of Reviews"].values)[:3]]

    result_text = f"""
Average price per neighbourhood:
{avg_price}

Top 3 listings by reviews:
{top_reviews[['Name','Price','Number of Reviews']]}

Average stay length: {avg_stay}
"""
    status_label.config(text=result_text)

# Buttons
tk.Button(root, text="Add Listing", command=add_listing).grid(row=len(fields), column=0)
tk.Button(root, text="Show Results", command=show_results).grid(row=len(fields), column=1)

status_label = tk.Label(root, text="")
status_label.grid(row=len(fields)+1, column=0, columnspan=2)

root.mainloop()
```
